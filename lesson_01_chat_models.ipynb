# Lesson 01 — Chat Models

## Goal

Learn how LangChain represents conversations using Chat Models and Message objects.

```python
import os
from dotenv import load_dotenv # Loads the .env file into the environment
from langchain_google_genai import ChatGoogleGenerativeAI
# Belongs to LangChain, this is LangChain integration package for google's gemini model 
```

```python
# Load the environment
load_dotenv() # Loads variables from .env file into the process environment
```

```python
# Read the model name
MODEL_NAME = os.environ["GEMINI_MODEL"]
API_KEY = os.environ["GOOGLE_GENERATIVE_AI_API_KEY"]
```

```python
# Create LangChain chat model
llm = ChatGoogleGenerativeAI(model = MODEL_NAME, temperature = 0, api_key = API_KEY) 
"""
LangChain automatically looks for GOOGLE_GENERATIVE_AI_API_KEY if exists, authentication succeeds
"""
```

```python
# Invoke the model

response = llm.invoke(
    "Explain what LangChain is in one paragraph."
)

print(response.content)
```

```python
from langchain_core.messages import HumanMessage

response = llm.invoke([
    HumanMessage(content="Tell me a joke.")
])
```

```python
message = HumanMessage(
    content = "Explain LangChain."
)

print(message)
```

```python
from langchain_core.messages import SystemMessage

system_message = SystemMessage(
    content="You are s senior Python instructor."
)

print(system_message)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

message = [
    SystemMessage(
        content="You are a senior Python instructor. Keep explanations concise and use simple examples"
    ),
    HumanMessage(
        content="Explain decorators"
    )
]

response = llm.invoke(message)
print(response)
```
